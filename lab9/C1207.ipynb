## Sampling

В ходе работы был рассмотрен равновероятностный сэмлинг и сэмплинг взевешенной выборки. Для взевешенной выборки были рассмотренны 2 случая: с линейной и квадратичной зависимостью веса от значения элемента.

Все три сэмплинга запускались для следующих значений n (число элементов в потоке) и m (размер сэмпла): 100 / 5, 1000 / 10, 1000 / 20, 10000 / 100. Число повторений r - 2000.

Как можно видеть из диаграмм, для равновернятностного сэмплинга число раз быть выбранным в сэмпл приблизительно одинаково для всех элементов и примерно равно r * (m / n).

Из диаграмм для взвешенных последовательностей можно заметить, что число раз быть выбранным за r пвотрений с ростом веса с линейной зависимостью возрастает также линейно, а с квадратичной - квадратично.

Исходный код: ...

```python
import matplotlib.pyplot as plt
%matplotlib inline

def plot_build(file):
    with open (file) as f:
        sample = [int(x) for x in str.split(f.read())]
    stream = list(range(1, len(sample) + 1))
    plt.figure(figsize = (16, 9))
    plt.xlim([0, len(stream) + 2])
    plt.grid(True)
    width = 1/1
    plt.ylabel('Entries count')
    plt.xlabel('Stream element')
    plt.bar(stream, sample, width, color="magenta")
```

## Равновероятностный сэмпл

### 1. n = 100, m = 5

```python
plot_build("sample100-5.txt")
```

### 2. n = 1000, m = 10

```python
plot_build("sample1000-10.txt")
```

### 3. n = 1000, m = 20

```python
plot_build("sample1000-20.txt")
```

### 4. n = 10000, m = 100

```python
plot_build("sample10000-100.txt")
```
